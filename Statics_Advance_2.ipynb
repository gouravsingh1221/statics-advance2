#Q1 What is hypothesis testing in statistics ?

<br>

Ans.    **Hypothesis testing** is a statistical method used to decide whether there is enough evidence in sample data to support or reject a claim about a population.

**Key Steps:**
1. **State the hypotheses:**
   - **Null hypothesis (H₀):** Assumes there is no effect or no difference.
   - **Alternative hypothesis (H₁):** Assumes there is an effect or difference.
2. **Choose a significance level (α):**
   - Common values are 0.05 or 0.01.
3. **Select a statistical test** based on data type and distribution.
4. **Calculate the test statistic** and find the p-value.
5. **Make a decision:**
   - If p-value ≤ α → Reject H₀ (evidence supports H₁).
   - If p-value > α → Fail to reject H₀ (not enough evidence).

**Example:**
A company claims its average delivery time is 3 days.  
You collect sample data and perform a hypothesis test to check if the actual average differs from 3 days.

#Q2  What is the null hypothesis, and how does it differ from the alternative hypothesis ?

<br>

Ans.  The **null hypothesis (H₀)** is the starting assumption in hypothesis testing that there is **no effect, no difference, or no relationship** in the population.  
It represents the default belief we test against.

The **alternative hypothesis (H₁ or Ha)** is the statement that **contradicts the null hypothesis**, claiming there is an effect, difference, or relationship.

**Key Differences:**
- **H₀ (Null Hypothesis):** Assumes nothing has changed or no relationship exists.
- **H₁ (Alternative Hypothesis):** Suggests there is a change or relationship.
- In testing, we either **reject H₀** in favor of H₁ or **fail to reject H₀** (we do not “accept” H₀, we just don’t have enough evidence against it).

**Example:**
- **H₀:** The average exam score of students is 70.
- **H₁:** The average exam score of students is not 70.

#Q3  Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

<br>

Ans. The **significance level (α)** is the probability threshold used in hypothesis testing to decide whether to reject the null hypothesis (H₀).  
It represents the maximum risk we are willing to take of making a **Type I error** — rejecting H₀ when it is actually true.

**Common values:** 0.05 (5%), 0.01 (1%).

**Role in testing:**
- Compare the **p-value** (calculated from sample data) to α:
  - If **p-value ≤ α** → Reject H₀ (evidence supports the alternative hypothesis).
  - If **p-value > α** → Fail to reject H₀ (not enough evidence against it).

**Example:**
If α = 0.05, we are willing to take a 5% chance of being wrong when rejecting the null hypothesis.

#Q4 What are Type I and Type II errors? Give examples of each.

<br>

Ans. In hypothesis testing, errors can occur when making decisions based on sample data.

1. **Type I Error (False Positive):**
   - Rejecting the null hypothesis (H₀) when it is actually true.
   - Probability of making this error is the **significance level (α)**.
   - **Example:** Concluding a new medicine works when it actually does not.

2. **Type II Error (False Negative):**
   - Failing to reject the null hypothesis when it is actually false.
   - Probability of making this error is **β** (related to the power of the test).
   - **Example:** Concluding a new medicine does not work when it actually does.

**Summary Table:**

| Decision / Reality     | H₀ True                  | H₀ False                 |
|------------------------|--------------------------|--------------------------|
| Reject H₀              | Type I Error (False +)   | Correct Decision         |
| Fail to Reject H₀      | Correct Decision         | Type II Error (False -)  |

#Q5 What is the difference between a Z-test and a T-test? Explain when to use each.

<br>

Ans. Both **Z-test** and **T-test** are statistical tests used to compare sample data with population parameters or compare two samples.  
The main difference lies in the sample size and whether the population standard deviation is known.

**1. Z-test:**
- Used when the **population standard deviation (σ)** is known.
- Usually applied when the sample size is **large (n ≥ 30)**.
- Assumes the data is normally distributed.
- **Example:** Testing if the average height of 1000 students differs from the known population mean.

**2. T-test:**
- Used when the **population standard deviation is unknown**.
- Common for **small sample sizes (n < 30)**.
- Uses the sample standard deviation (s) instead of σ.
- **Example:** Comparing the average exam score of 20 students to a claimed mean.

**Summary Table:**

| Feature                  | Z-test                               | T-test                                 |
|--------------------------|---------------------------------------|-----------------------------------------|
| Population σ known?      | Yes                                  | No                                     |
| Sample size              | Large (≥ 30)                         | Small (< 30)                           |
| Distribution used        | Standard Normal Distribution (Z)     | Student’s t-distribution               |

###Q6 Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10  # number of trials
p = 0.5  # probability of success
size = 1000  # number of samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=np.arange(data.min(), data.max()+2)-0.5, color='skyblue', edgecolor='black')
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

###Q7 mplement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu_0 = 50

# Known population standard deviation (assumed for Z-test)
sigma = 0.5

# Sample statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Z-statistic formula
z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Z-Statistic:", z_stat)
print("P-Value:", p_value)

# Decision at 5% significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Evidence suggests mean is different from", mu_0)
else:
    print("Fail to reject the null hypothesis: No significant difference from", mu_0)
```

###Q8  Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=100, scale=15, size=200)

# Calculate sample mean and standard error
mean_value = np.mean(data)
std_err = stats.sem(data)  # standard error of the mean

# Confidence interval at 95%
confidence = 0.95
ci_lower, ci_upper = stats.t.interval(confidence, len(data)-1, loc=mean_value, scale=std_err)

# Print results
print("Sample Mean:", mean_value)
print("95% Confidence Interval: ({:.2f}, {:.2f})".format(ci_lower, ci_upper))

# Plot histogram of data
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title("Normal Distribution Simulation")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

###Q9 Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Function to calculate Z-scores and plot histogram
def calculate_and_plot_zscores(data):
    # Calculate Z-scores
    z_scores = stats.zscore(data)

    # Display first few Z-scores
    print("First 5 Z-scores:", z_scores[:5])

    # Plot histogram of Z-scores
    plt.hist(z_scores, bins=20, color='lightgreen', edgecolor='black')
    plt.title("Histogram of Z-scores")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    return z_scores

# Example dataset
data = [12, 15, 14, 10, 8, 10, 12, 14, 16, 18, 20, 22, 21, 19, 17, 15]

# Call the function
z_scores = calculate_and_plot_zscores(data)
```
